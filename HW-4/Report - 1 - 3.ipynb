```python
import numpy
import networkx


def generate_scale_free_network(N, y):
    sequence = numpy.random.zipf(y, N)
    while not networkx.is_graphical(sequence):
        sequence = numpy.random.zipf(y, N)

    graph = networkx.configuration_model(sequence)
    simple_graph = networkx.Graph(graph)
    multi_edges = graph.number_of_edges() - simple_graph.number_of_edges()
    self_loops_percentage = (len(list(networkx.selfloop_edges(graph)))) / graph.number_of_edges()
    multi_edges_percentage = multi_edges / graph.number_of_edges()
    print(f'% of self-loops: {self_loops_percentage}')
    print(f'% of multi-edges: {multi_edges_percentage}')
    return [self_loops_percentage, multi_edges_percentage]
```

```python
generate_scale_free_network(1000, 2.2)
```

```python
N = numpy.arange(100000, 1000000, 100000).tolist()
self_edge_list = []
multi_link_list = []
for n in N:
    self_edge, multi_link = generate_scale_free_network(n, 2.2)
    self_edge_list.append(self_edge)
    multi_link_list.append(multi_link)
```

```python
from matplotlib import pyplot as plt

plt.plot(N, self_edge_list)
plt.show()
```

```python
plt.plot(N, multi_link_list)
plt.show()
```
