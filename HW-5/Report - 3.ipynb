```python
def import_data(path):
    node = []
    degree = []
    eigen = []
    betweenness = []

    with open(path) as file:
        for line in file:
            n, d, e, b = line.split()
            node.append(str(n))
            degree.append(float(d))
            eigen.append(float(e))
            betweenness.append(float(b))

    return node, degree, eigen, betweenness
```

```python
from matplotlib import pyplot as plt


def plot_centrality(node, degree, eigen, between):
    plt.figure(figsize=(35, 5))

    plt.subplot(1, 3, 1)
    plt.bar(node, degree, color='m')
    plt.title('Degree Centrality')
    plt.xlabel('Node')
    plt.ylabel('Centrality')
    plt.grid()

    plt.subplot(1, 3, 2)
    plt.bar(node, eigen, color='c')
    plt.title('Eigen Centrality')
    plt.xlabel('Node')
    plt.ylabel('Centrality')
    plt.grid()

    plt.subplot(1, 3, 3)
    plt.bar(node, between, color='g')
    plt.title('Betweenness Centrality')
    plt.xlabel('Node')
    plt.ylabel('Centrality')
    plt.grid()

    plt.show()
```

```python
node, degree, eigen, betweenness = import_data('Data/unnormalized.txt')
plot_centrality(node, degree, eigen, betweenness)
```

```python
node, degree, eigen, betweenness = import_data('Data/normalized.txt')
plot_centrality(node, degree, eigen, betweenness)
```
