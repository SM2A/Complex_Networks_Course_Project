در این بخش فایل داده ورودی خط به خط خوانده شده و با توجه به دو سر یال، ارتباط آن ها در ماتریس مجاورتی که به تدریح در حال تکمیل است اضافه شده است.

سپس در ادامه به ازای هر گره به کمک کتابخانه $numpy$، تعداد خانه های غیر صفر مرتبط با گره را می شماریم و نتیجه آن را به لیست مربوطه اضافه میکنیم.

```python
NO_LINK = 0
LINK = 1

adj_mat = []

with open('Data/Twitter_edges.txt') as file:
    for line in file:
        start, end = line.split()
        start, end = int(start), int(end)
        max_size = int(max(start, end))
        while len(adj_mat) < max_size + 1:
            adj_mat.append([])
        for row in adj_mat:
            while len(row) < max_size + 1:
                row.append(NO_LINK)
        adj_mat[start][end] = LINK
        adj_mat[end][start] = LINK
```

```python
import numpy
from matplotlib import pyplot as plt

degree = []
for node in range(len(adj_mat)):
    adj  = numpy.array(adj_mat[node])
    degree.append(len((adj == 1).nonzero()[-1]))
    
plt.figure(figsize=(10,10))

plt.subplot(2, 2, 1)
plt.hist(degree, bins=len(degree), color='c')
plt.title('Degree Distribution')
plt.xlabel('Degree')
plt.ylabel('Count')

plt.subplot(2, 2, 2)
plt.hist(degree, bins=len(degree), color='c')
plt.title('Degree Distribution')
plt.xlabel('Degree')
plt.ylabel('Count')
plt.xlim([0, 100])

plt.subplot(2, 2, 3)
plt.hist(degree, bins=len(degree), color='c')
plt.title('Degree Distribution')
plt.xlabel('Degree')
plt.ylabel('Count')
plt.xlim([0, 50])

plt.subplot(2, 2, 4)
plt.hist(degree, bins=len(degree), color='c')
plt.title('Degree Distribution')
plt.xlabel('Degree')
plt.ylabel('Count')
plt.xlim([0, 25])

plt.tight_layout()
plt.show()
```
