```python
import numpy
from matplotlib import pyplot as plt

data = numpy.loadtxt('Data/cities.txt', dtype=int)
hist, bins = numpy.histogram(data, bins=sorted(list(set(data))), density=True)

plt.figure(figsize=(30, 15))

plt.subplot(1, 2, 1)
plt.title('PDF')
plt.xlabel('Population')
plt.ylabel('Probability')
plt.plot(bins[:-1], hist, 'go')

plt.subplot(1, 2, 2)
plt.title('PDF')
plt.xlabel('Population')
plt.ylabel('Probability')
plt.xscale('log')
plt.yscale('log')
plt.plot(bins[:-1], hist, 'go')

plt.show()
```

```python
cdf = numpy.cumsum(hist)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('CDF')
plt.xlabel('Population')
plt.ylabel('Cumulative Probability')
plt.plot(bins[:-1], cdf, 'r')

plt.subplot(1, 2, 2)
plt.title('CDF')
plt.xlabel('Population')
plt.ylabel('Cumulative Probability')
plt.xscale('log')
plt.yscale('log')
plt.plot(bins[:-1], cdf, 'r')

plt.show()
```
